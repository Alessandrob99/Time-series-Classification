```python
import numpy as np
import statsmodels.api as sm
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt 
from statsmodels.tsa.stattools import adfuller


lowess = sm.nonparametric.lowess

plt.rcParams.update({'figure.figsize':(40,20), 'figure.dpi':400})




df = pd.read_csv('C:/Users/aless/Desktop/AMZN.csv',header=0)
df = df[df.Date.str.split('-',expand=True)[0].isin(["2020","2021","2022"])]



fig, axes = plt.subplots(nrows=2, ncols=2, sharex=False)
axes[0,0].plot(df.Open)
axes[0,0].set_title("Serie Originale")

z = lowess(df.Open,np.arange(0,len(df.Open),1),frac=1./25)
axes[1,0].scatter(z[:,0],z[:,1])
axes[1,0].set_title("LOWESS")
```
