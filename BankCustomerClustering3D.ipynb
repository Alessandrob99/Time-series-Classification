```python

%matplotlib qt

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd
from sklearn.preprocessing import normalize
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import MaxNLocator
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from numpy import concatenate
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.manifold import TSNE

S = pd.read_csv('C:/Users/aless/Desktop/BankCustomerData.csv')

S['job']=S['job'].astype('category').cat.codes
S['marital']=S['marital'].astype('category').cat.codes
S['education']=S['education'].astype('category').cat.codes
S['default']=S['default'].astype('category').cat.codes
S['housing']=S['housing'].astype('category').cat.codes
S['loan']=S['loan'].astype('category').cat.codes
S['contact']=S['contact'].astype('category').cat.codes
S['month']=S['month'].astype('category').cat.codes
S['poutcome']=S['poutcome'].astype('category').cat.codes
S['term_deposit']=S['term_deposit'].astype('category').cat.codes



"""
#PCA
pca = PCA(n_components=2)
pca.fit(S)
S = PCA.transform(pca,S)

"""
S = S[["age","balance","duration"]] 

S = S.loc[np.random.permutation(S.index)[:20000]]  #SUBSAMPLE da 42k a 20k

S = normalize(S, norm='l2', axis=1, copy=True, return_norm=False) 

kmeans = KMeans(n_clusters=4,init='k-means++')
#teoricamente consideriamo feature diverse in fase di fit ma non cambia nulla
#kmeans.n_features_in_ = 2
#kmeans.feature_names_in_ = ["age","duration"]

labels = kmeans.fit_predict(S)
#selezioamo i vari cluster

filtered_label0 = S[labels == 0]
 
filtered_label1 = S[labels == 1]

filtered_label2 = S[labels == 2]

filtered_label3 = S[labels == 3]


fig = plt.figure()
kplot = fig.add_subplot(111, projection='3d')





# Data for three-dimensional scattered points
kplot.scatter(filtered_label0[:,0], filtered_label0[:,1], filtered_label0[:,2], c='red', label = 'Cluster 1')
kplot.scatter(filtered_label1[:,0],filtered_label1[:,1],filtered_label1[:,2],c ='green', label = 'Cluster 2')
kplot.scatter(filtered_label2[:,0],filtered_label2[:,1],filtered_label2[:,2],c ='black', label = 'Cluster 2')
kplot.scatter(filtered_label3[:,0],filtered_label3[:,1],filtered_label3[:,2],c ='yellow', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'indigo', s = 200)
kplot.set(xlabel="age", ylabel ="balance", zlabel = "duration")


plt.legend()
plt.rcParams["figure.figsize"] = (10,10)
plt.title("Kmeans")
plt.show()
```
